```python
direct = "C:\\Users\\fabia\\Desktop\\db\\athletics\\"
direct1 = "C:\\Users\\fabia\\Desktop\\db\\cricket\\"
direct2 = "C:\\Users\\fabia\\Desktop\\db\\football\\"
direct3 = "C:\\Users\\fabia\\Desktop\\db\\rugby\\"
direct4 = "C:\\Users\\fabia\\Desktop\\db\\tennis\\"

NewsAthletics = os.listdir(direct)
NewsCricket = os.listdir(direct1)
NewsFootball = os.listdir(direct2)
NewsRugby = os.listdir(direct3)
NewsTennis = os.listdir(direct4)
```

```python
import pandas as pd
```

```python
res = pd.DataFrame(columns=('Type', 'Title', 'News'))
```

```python
rows=[]
columns=['Type','Title','News']
for new in NewsAthletics:
    f = open(direct+new)
    mensaje=f.readline()
    mensaje2=f.read()
    row=['athletics',mensaje.strip(),mensaje2.strip()]
    rows.append(row)
    f.close

#rows=[]
columns=['Type','Title','News']
for new in NewsCricket:
    f = open(direct1+new)
    mensaje=f.readline()
    mensaje2=f.read()
    row=['cricket',mensaje.strip(),mensaje2.strip()]
    rows.append(row)
    f.close
    
#rows=[]
columns=['Type','Title','News']
for new in NewsFootball:
    f = open(direct2+new)
    mensaje=f.readline()
    mensaje2=f.read()
    row=['football',mensaje.strip(),mensaje2.strip()]
    rows.append(row)
    f.close
    
#rows=[]
columns=['Type','Title','News']
for new in NewsRugby:
    f = open(direct3+new)
    mensaje=f.readline()
    mensaje2=f.read()
    row=['rugby',mensaje.strip(),mensaje2.strip()]
    rows.append(row)
    f.close
    

columns=['Type','Title','News']
for new in NewsTennis:
    f = open(direct4+new)
    mensaje=f.readline()
    mensaje2=f.read()
    row=['tennis',mensaje.strip(),mensaje2.strip()]
    rows.append(row)
    f.close
    
```

```python
df=pd.DataFrame(rows,columns=columns)
```
